# Day 2: Password Philosophy
## Setup
First whe need loads the input data.

```python
def loadInput() -> list:
    file = open("input.txt", "r")
    input = []
    for line in file.readlines():
        text = line.split()
        numbers = text[0].split("-")
        character = text[1][:1]
        password = text[2]
        input.append((int(numbers[0]), int(numbers[1]), character, password))
    return input
```

```python
inputData = loadInput()
sampleData = [
    [1, 3, 'a', 'abcde'], 
    [1, 3, 'b', 'cdefg'], 
    [2, 9, 'c', 'ccccccccc'], 
]
```

## Let's start to code!!

This solution it's good, but i think it's possible simplify the if statements

```python
def firstTry(policies: list) -> int:
    count = 0

    for policie in policies:
        first = policie[0]
        second = policie[1]
        character = policie[2]
        password = policie[3]

        if password[first - 1] == character and password[second - 1] != character:
            count += 1
            
        if password[first - 1] != character and password[second - 1] == character:
            count += 1

    return count
```

```python
solution = firstTry(inputData)
print(f"Solution: {solution}")
```

Using `XOR` it's possible simplified the if statements

```python
def secondTry(policies: list) -> int:
    count = 0

    for policie in policies:
        first = policie[0]
        second = policie[1]
        character = policie[2]
        password = policie[3]

        if (password[first - 1] == character) ^ (password[second - 1] == character):
            count += 1

    return count
```

```python
solution = secondTry(inputData)
print(f"Solution: {solution}")
```

## Test
It looks good, but we need make a little test to ensure that the answer it's correct.

```python
assert(firstTry(sampleData) == 1)
```

## Performance
The second solution has been executed 1.5x faster.

```python
%timeit -r 10 -n 5 firstTry(inputData)
%timeit -r 10 -n 5 secondTry(inputData)
```
